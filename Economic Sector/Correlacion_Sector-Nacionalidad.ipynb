# Correlacion nacionalidad-sector

```python
#importing the libraries.
import pandas as pd
```

## Defining the data frames used

```python
# Load the data into a DataFrame
data = pd.read_csv("Poblacion en cada sector.csv", sep=";")

# Extract column names
sectors = [ "Primario", "Industria", "Construccion", "Servicios"]
data
```

```python
# Calculate mean and standard deviation for each sector
data[sectors] =  data[sectors].div(data["Total"], axis=0)
data
```

```python
mean_sectors = data[sectors].mean()
std_sectors = data[sectors].std()

# Calculate the normalized value for each municipio and sector
df_Sectores = data.copy()
for sector in sectors:
    df_Sectores[sector] = (data[sector] - mean_sectors[sector]) / std_sectors[sector]

df_Sectores = df_Sectores.drop('Total', axis=1)
df_Sectores = df_Sectores.drop(index=0)
df_Sectores = df_Sectores.drop('No Consta', axis=1)
    
df_Sectores
```

```python
# Load the file
file_path = "Poblacion de cada nacionalidad.csv"  # Replace with your file path
data2 = pd.read_csv(file_path, sep=";", encoding="latin-1")

# 'Municipio' column will be ignored because it's not numeric
data2_numeric = data2.drop(columns=["Municipio"]).apply(pd.to_numeric, errors='coerce')

# Filter columns where the value in the first row (index 0) is less than 100
columns_to_drop = [col for col in data2_numeric.columns if data2_numeric[col].iloc[0] < 100]

# Drop the selected columns
data2 = data2.drop(columns=columns_to_drop)


# Extract the first row (columns serve as keys)
keys = [col for col in data2.columns if col not in ["Total", "Municipio"]]
data2[keys] = data2[keys].div(data2["Total"], axis=0)

data2 = data2.drop('Total', axis=1)
data2 = data2.drop(index=0)
data2 = data2.drop('Otros (Asia)', axis=1)
data2.rename(columns={"No aplicable (siempre ha vivido en España)": "España"}, inplace=True)

# Extract the first row (columns serve as keys)
keys = [col for col in data2.columns if col not in ["Total", "Municipio"]]

data2
```

```python
mean_nacionality = data2[keys].mean()
std_nacionality = data2[keys].std()

# Calculate the normalized value for each municipio and sector
df_Nacionalidad = data2.copy()
for Nacionalidad in keys:
    df_Nacionalidad[Nacionalidad] = (data2[Nacionalidad] - mean_nacionality[Nacionalidad]) / std_nacionality[Nacionalidad]

df_Nacionalidad
```

## Studying the correlation

```python
#Find the common municipalities between both data-frames
municipios = set(df_Nacionalidad["Municipio"]).intersection(df_Sectores["Municipio"])

df_Nacionalidad = df_Nacionalidad[df_Nacionalidad["Municipio"].isin(municipios)].set_index("Municipio")
df_Sectores = df_Sectores[df_Sectores["Municipio"].isin(municipios)].set_index("Municipio")
```

```python
# Step 2: Multiply Each Nationality by Each Sector
result = pd.DataFrame({
    (nat, sec):df_Nacionalidad[nat] * df_Sectores[sec]
    for nat in df_Nacionalidad.columns
    for sec in df_Sectores.columns
})

Correlacion_prom = result.mean()
Df_Correlacion = Correlacion_prom.unstack()
Df_Correlacion
```

```python
Df_Correlacion.to_csv('Df_Correlacion.csv', sep=';', index=True)
```
