# Filter the number of immigrants by age (working - not working) 

```python
#importing the libraries.
import pandas as pd
```

```python
# Load the data into a DataFrame
df = pd.read_csv("Poblacion inmigrante.csv", sep=";")

df
```

```python
#Select those rows with age 16-64
df = df[df['Edad'] == "16-64"]
df = df.drop('Edad', axis=1)

df
```

```python
df.to_csv("Poblacion de cada nacionalidad.csv", sep=';', index=True)
```
