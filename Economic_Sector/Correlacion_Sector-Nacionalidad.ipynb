# Correlacion nacionalidad-sector

```python
#importing the libraries.
import pandas as pd
```

```python
# Load the data into a DataFrame
data = pd.read_csv("Poblacion en cada sector.csv", sep=";")

# Extract column names
sectors = [ "Primario", "Industria", "Construccion", "Servicios"]
data
```

```python
# Calculate mean and standard deviation for each sector
data[sectors] =  data[sectors].div(data["Total"], axis=0)
data[sectors]
```

```python
mean_sectors = data[sectors].mean()
std_sectors = data[sectors].std()

# Calculate the normalized value for each municipio and sector
normalized_data = data.copy()
for sector in sectors:
    normalized_data[sector] = (data[sector] - mean_sectors[sector]) / std_sectors[sector]

data_dict = data.set_index("Municipio").to_dict(orient="index")

    
normalized_data[sectors]
```
